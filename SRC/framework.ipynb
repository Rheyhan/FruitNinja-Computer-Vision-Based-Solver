```python
from ultralytics import YOLO
import cv2
import numpy as np
import time
import os
import argparse
import supervision as sv
from mss import mss
```

```python
model = YOLO("bestModel.pt")
model.to("cuda")
```

```python
def startAutoSlicer(visualize=True):
    with mss() as sct:
        fps_counter = 0
        start_time = time.time()

        while True:
            
            mon = sct.monitors[2]
            img = np.asarray(sct.grab(mon))   
            img = cv2.resize(img, (640, 640))
            img = cv2.cvtColor(img, cv2.COLOR_BGRA2BGR)
            results = model.predict(img, verbose=True, conf=0.7, iou=0.5)

            if visualize:
                annotated_image = results[0].plot()
                
                # Calculate FPS
                fps_counter += 1
                elapsed_time = time.time() - start_time
                fps = fps_counter / elapsed_time
                
                # Add FPS text to the image
                cv2.putText(annotated_image, f'FPS: {fps:.1f}', (10, 30), 
                        cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
                
                cv2.imshow("screenshot", annotated_image)
                if cv2.waitKey(1) & 0xFF == ord("q"):
                    cv2.destroyAllWindows()
                    break
```

```python
startAutoSlicer()
```
