```python
from roboflow import Roboflow
import shutil
import os
import tqdm
```

```python
import json

with open("../API_Key.json", "r") as f:
    keys = json.load(f)

rf = Roboflow(api_key=keys["ROBOFLOW_API_KEY"])
```

```python
def unpack_move(src, dst_dirname):
    dst_path = os.path.join("../DATA/RAW/", dst_dirname)
    dst_path_img = os.path.join(dst_path, "images")
    dst_path_label = os.path.join(dst_path, "labels")
    os.makedirs(dst_path_img, exist_ok=True)
    os.makedirs(dst_path_label, exist_ok=True)

    src_dirname = src.split("\\")[-1]
    index = 0
    shutil.move(os.path.join(src_dirname, "data.yaml"), os.path.join(dst_path, "data.yaml"))
    for i in os.listdir(src_dirname):
        if os.path.isdir(os.path.join(src_dirname, i)):
            img_dir = os.path.join(src_dirname, i, "images")
            label_dir = os.path.join(src_dirname, i, "labels")
            for img_file in tqdm.tqdm(os.listdir(img_dir)):
                if img_file.endswith(".jpg") or img_file.endswith(".jpeg") or img_file.endswith(".png"):

                    shutil.move(os.path.join(img_dir, img_file), os.path.join(dst_path_img, f"{index:04d}.jpg"))
                    label_file = img_file.replace(".jpg", ".txt")
                    shutil.move(os.path.join(label_dir, label_file), os.path.join(dst_path_label, f"{index:04d}.txt"))
                    index += 1

    shutil.rmtree(src_dirname)
```

```python
# 1
# https://universe.roboflow.com/pythonai-hf-lsuae/fruit-ninja-2noom/dataset/3
project = rf.workspace("pythonai-hf-lsuae").project("fruit-ninja-2noom")
version = project.version(3)
dataset = version.download("yolov11")

unpack_move(dataset.location, "DATA 1")
```
